# üìñ Pandas Library

**Pandas** is an open-source Python library used for **data analysis and manipulation**.  
It provides easy-to-use data structures and tools to work with structured data (like tables or spreadsheets).  

The two main data structures in pandas are:  
- **Series** ‚Üí A one-dimensional labeled array (like a single column in Excel).  
- **DataFrame** ‚Üí A two-dimensional labeled data structure (like an entire Excel sheet with rows and columns).  

Official Docs: https://pandas.pydata.org/docs/

## Drop Duplicates

drop_duplicates() is used to remove duplicate rows from a DataFrame (or duplicate values from a Series).
By default, it keeps the first occurrence of each duplicate and drops the rest.
You can change this behavior with the keep parameter:

keep='first' ‚Üí keeps the first duplicate (default).

keep='last' ‚Üí keeps the last duplicate.

keep=False ‚Üí removes all duplicates.

You can also specify a subset of columns with the subset parameter to check for duplicates only in those columns.

````python
data.drop_duplicates(subset=['city'])
#drops the reoccuring values
````

````python
data.drop_duplicates(subset=['city', 'season'])
````

````python
data.drop_duplicates('season', keep='last')[['season', 'winner']].sort_values('season')
````

## isin() 
isin() function in pandas is used to filter rows by checking whether each value in a column (or DataFrame) matches any value from a given list.

üîπ Simple Explanation

Think of it like asking: ‚ÄúIs this value in my list of allowed values?‚Äù

It returns True/False for each row, which you can use for filtering.

````python
#isin()
#Find the most destructive death over batsman in the history of ipl
#strike rate = (num of runs/ num of balls)/100
#min batsman 200 balls in over 16-20
````

## Pivot Table
A pivot table in pandas is a way to summarize and reorganize data by grouping it on one or more keys and applying an aggregation (like sum, mean, count, etc.).

üîπ Simple Definition

A pivot table lets you turn columns into rows (and vice versa) and quickly see summaries, just like Excel pivot tables.

Key Parameters:

    data: The DataFrame from which to create the pivot table.
    values: Column(s) to aggregate. If not specified, all numeric columns will be aggregated.
    index: Column(s) to use for the new DataFrame's index (rows). Can be a single column name or a list of column names.
    columns: Column(s) to use for the new DataFrame's columns. Can be a single column name or a list of column names.
    aggfunc: Aggregation function(s) to use. Can be a single function (e.g., 'mean', 'sum', np.mean), a list of functions, or a dictionary mapping columns to functions.
        Common strings: 'mean', 'sum', 'count', 'min', 'max', 'median', 'first', 'last'.
        NumPy functions: np.mean, np.sum, np.median, etc.
    fill_value: Value to replace missing values (NaN) in the pivot table output. Useful for making the table look cleaner or for further calculations.
    margins: Boolean (default False). If True, adds row and column subtotals/totals (called "margins").
    dropna: Boolean (default True). If True, don't include columns whose entries are all NaN.
    margins_name: String (default 'All'). Name of the row/column that will contain the totals when margins=True.

````python
#Pivot table makes data more readabale
# fetch average amount of money spend on burgers by male on kolkata on daily basis
food=pd.read_csv('food_spends.csv')
food.head()
````

## Missing Values


### dropna()

dropna() removes missing values (NaN) from a DataFrame or Series.

**General Structure:**
```python
DataFrame.dropna(
    axis=0,       # 0 = drop rows, 1 = drop columns
    how='any',    # 'any' = drop if ANY NaN, 'all' = drop if ALL NaN
    thresh=None,  # minimum number of non-NaN values required to keep
    subset=None,  # apply drop only on specific column(s)
    inplace=False # whether to modify the DataFrame directly
)

### fillna()

Replaces missing values (NaN) with a specified value or method.

**General Syntax:**

```python
DataFrame.fillna(
    value=None,    # scalar, dict, or Series to replace NaN
    method=None,   # 'ffill' = forward fill, 'bfill' = backward fill
    axis=None,     # 0 = fill by rows, 1 = fill by columns
    inplace=False, # modify DataFrame directly if True
    limit=None     # max number of NaNs to fill
)
