## -Matplotlib

Matplotlib is a popular data visualization library in Python.
It provides functions and tools to create a wide variety of static, interactive, and animated plots.
Think of it as Python’s "drawing tool" for turning raw data into graphs, charts, and figures so that patterns, trends, and insights can be seen clearly.

It was originally developed by John D. Hunter in 2003 and is now one of the most widely used visualization libraries in the Python ecosystem.

### What Matplotlib Does

- Matplotlib allows you to:
- Plot Different Types of Graphs
- Line plots
- Bar charts
- Histograms
- Scatter plots
- Pie charts
- Area plots
- 3D plots (with mplot3d)
- Customize Plots Fully
- Change titles, labels, axes, colors, line styles, markers, legends, and fonts.
- Control figure size and resolution.
- Annotate specific points on graphs.
- Work with Other Libraries
- Integrates seamlessly with NumPy, Pandas, and SciPy.
- Often used in combination with pandas DataFrames for quick plotting of datasets.
- Support for Different Backends
- Can render plots in Jupyter Notebook, Python scripts, or even export them as images (PNG, PDF, SVG, etc.).

### Why Matplotlib is Important?

Data Science & Machine Learning: Used for visualizing trends and model outputs.
Research & Engineering: Helps in plotting scientific/engineering data.
Education: Great for teaching math, statistics, and data analysis.
Customization: Gives very fine control compared to other libraries (like Seaborn).

## Plotting Functions
| Function                      | Definition                                           | Common Parameters                                      |
|-------------------------------|------------------------------------------------------|--------------------------------------------------------|
| `plt.plot(x, y)`              | Creates a 2D line plot.                              | color, linestyle, linewidth, marker, label             |
| `plt.scatter(x, y)`           | Creates a scatter plot.                              | s (size), c (color), marker, alpha, cmap, label        |
| `plt.bar(x, height)`          | Draws a vertical bar chart.                          | width, color, align, label                             |
| `plt.barh(y, width)`          | Draws a horizontal bar chart.                        | height, color, align, label                            |
| `plt.hist(data, bins)`        | Creates a histogram.                                 | bins, color, alpha, edgecolor                          |
| `plt.pie(sizes)`              | Creates a pie chart.                                 | labels, colors, autopct, startangle, shadow, explode   |
| `plt.stem(x, y)`              | Creates a stem (lollipop-style) plot.                | linefmt, markerfmt, basefmt                            |
| `plt.step(x, y)`              | Creates a step plot.                                 | where (‘pre’, ‘post’, ‘mid’), color, linestyle         |
| `plt.fill(x, y)`              | Fills area under a curve.                            | color, alpha, label                                    |
| `plt.fill_between(x, y1, y2)` | Fills area between two curves.                       | color, alpha, where, label                             |

X → one variable

- for bivariate analysis
- numerical vs numerical
- use case- finding correlation

### What It Does

Shows relationships/trends (positive, negative, none)
Detects outliers (points far from the rest)
Finds clusters/groups in data
Can add color/size to show extra information

## Bar Chart in Matplotlib

### Definition
A **bar chart** is a plot that uses **rectangular bars** to represent data.  

- The **length/height of each bar** is proportional to the value it represents.  
- Bars can be **vertical** (`plt.bar()`) or **horizontal** (`plt.barh()`).  

It is mainly used to **compare categories** of data.  

### What Bar Charts Do
Bar charts are useful for:

- Comparing **different groups or categories** (e.g., sales of products, marks of students).  
- Showing **trends over time** (if categories are time-based).  
- Highlighting **highest/lowest values** clearly.  
- Representing **discrete/categorical data** (not continuous like line plots).  


````python
#simple bar chart
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors, children)
````

````python
#horicontal bar chart 
plt.barh(colors, children)
````

````python
#colors and labels
df = pd.read_csv('batsman_season_record.csv')
df
````

````python
plt.bar(np.arange(df.shape[0]) -0.2, df['2015'], width=0.2, color = 'pink')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color = 'purple')
plt.bar(np.arange(df.shape[0]) +0.2, df['2017'], width=0.2, color = 'yellow')
````

## Histogram in Matplotlib

### Definition
A **histogram** is a type of plot that shows the **distribution of numerical data** by dividing the data range into **intervals (bins)** and counting how many values fall into each bin.  

- X-axis → Represents the data intervals (bins).  
- Y-axis → Represents the frequency (number of data points in each bin).  
- Unlike a bar chart (which compares categories), a histogram is used for **continuous numerical data**.  

### What Histograms Do
Histograms are useful for:
- Understanding the **shape of data distribution** (normal, skewed, uniform, etc.).  
- Finding **patterns, trends, and outliers** in data.  
- Showing **frequency of values** within certain ranges.  
- Useful in **statistics, machine learning, and data analysis**.  

### Function in Matplotlib
- `plt.hist(data, bins, color, alpha, edgecolor)` → Creates a histogram.  

**Common Parameters:**  
- `data` → The dataset to be divided into bins.  
- `bins` → Number of intervals (default = 10).  
- `color` → Fill color of bars.  
- `alpha` → Transparency of bars (0 to 1).  
- `edgecolor` → Color of bin edges.  


````python
#simple data
data = [32, 45, 56, 10, 15, 27, 61]
bins = [10,25, 40,55,70]

plt.hist(data, bins)
````

### Logarithmic Scale
#### Definition

A logarithmic scale is a type of scale used in graphs where each tick mark on an axis increases by a factor (usually 10) instead of a fixed amount.
In a linear scale, equal distances mean equal differences (e.g., 1, 2, 3, 4…).
In a logarithmic scale, equal distances mean equal ratios (e.g., 1, 10, 100, 1000…).
So instead of growing additively, values grow multiplicatively.

### What It Does

Helps visualize data that spans several orders of magnitude (very small to very large numbers).
Makes exponential growth, decay, or power-law relationships easier to see.
Reduces skew when large values dominate the graph

````python
#logarithmic scale
arr = np.load('big-array.npy')
arr
````

````python
plt.hist(arr, bins=[10,20,30,40,50,60,70], log= True)
plt.show
````

## Pie Chart in Matplotlib

### Definition
A **Pie Chart** is a circular statistical graphic that is divided into slices to illustrate the **numerical proportion** of data.  
Each slice represents a **category's contribution** to the whole dataset.  

- The **whole circle = 100% of the data**.  
- Each **slice size = (category value / total value) × 360°**.  

### What a Pie Chart Does
- Shows the **relative percentage or proportion** of different categories.  
- Useful for comparing **parts of a whole**.  
- Makes it easy to see which categories are **dominant or minor contributors**.  
- Best used when the number of categories is small (2–7).  

### Function in Matplotlib
- `plt.pie(data, labels, colors, autopct, startangle, explode)` → Creates a pie chart.  

**Common Parameters:**  
- `data` → List/array of values to plot.  
- `labels` → Names for each slice.  
- `colors` → Slice colors.  
- `autopct` → Format string to display percentages (e.g., `"%.1f%%"`).  
- `startangle` → Starting rotation of the pie chart (default is 0).  
- `explode` → Offset a slice from the center to highlight it.  


````python
#simple data
data = [23,45,100,20,49]
subjects = ['physics', 'Compsci', 'english', 'maths', 'Urdu']
plt.pie(data, labels= subjects)
plt.show()
````
